# OpenClaw — Hướng dẫn khởi động

> Copy từng lệnh → dán vào **Terminal** bên dưới → **Enter**.
> Làm theo thứ tự từ trên xuống.

---
## Bước 1 — Kiểm tra cài đặt

Xác nhận `openclaw` đã sẵn sàng:

````bash
openclaw --version
````

> **Lỗi `command not found`?** Chạy lệnh này rồi thử lại:
> ```bash
> source ~/.bashrc
> ```

---
## Bước 2 — Onboard Wizard

Wizard dẫn bạn qua từng bước cấu hình. Trả lời câu hỏi rồi bấm **Enter**.
Mất khoảng **2–3 phút**.

````bash
openclaw onboard --install-daemon
````

### 2a. Kết nối AI

Wizard hỏi bạn muốn dùng AI nào. Chọn provider và dán API key khi được hỏi.

| Provider | Lấy API key ở đâu |
|----------|-------------------|
| **Anthropic** (Claude) | [console.anthropic.com](https://console.anthropic.com) |
| **OpenAI** (GPT) | [platform.openai.com](https://platform.openai.com) |
| **Ollama** *(local)* | Nhập URL `http://127.0.0.1:11434` |
| **Custom** | Bất kỳ endpoint tương thích OpenAI / Anthropic |

> Key được lưu trong `~/.openclaw/` — không cần file `.env`.

### 2b. Kết nối kênh chat

Wizard hỏi bạn muốn nhắn tin với AI qua app nào.
Có thể **bỏ qua** (Enter) và cài thêm sau.

| Kênh | Cần gì | Ghi chú |
|------|--------|---------|
| **Telegram** | Bot token từ [@BotFather](https://t.me/BotFather) | ✅ Dễ nhất, khuyến nghị |
| **WhatsApp** | Quét QR bằng điện thoại | Cần scan lại khi hết session |
| **Discord** | Bot token từ Discord Developer Portal | Hỗ trợ server + DM |
| **Signal** | signal-cli + số điện thoại | Bảo mật cao |
| **WebChat** | Không cần gì thêm | Dùng ngay qua Dashboard |

> **Chưa biết chọn gì?** Chọn **Telegram** — paste bot token là xong.

### 2c. Kiểm tra Gateway

Sau khi wizard hoàn tất, daemon đã được cài (systemd user unit).
Kiểm tra Gateway đang chạy:

````bash
openclaw gateway status
````

> **Gateway chưa chạy?** Khởi động thủ công:
> ```bash
> openclaw gateway run --port 18789
> ```

## Bước 3 — Mở Dashboard

Chạy lệnh sau để lấy **token** truy cập:

````bash
openclaw dashboard
````

Lệnh trên in ra URL dạng:
```
http://127.0.0.1:18789/?token=xxxxxxxx
```

> Trong Codespaces / Firebase Studio, `127.0.0.1` không mở được trực tiếp.  
> Làm theo các bước sau:

1. **Copy token** từ URL trên (phần sau `?token=`)
2. Mở tab **Ports** ở góc dưới màn hình
3. Tìm port **18789** → click biểu tượng **🌐** để mở Control UI
4. Trình duyệt mở URL dạng `https://<codespace>-18789.app.github.dev` → thêm `?token=<token>` vào cuối URL rồi nhấn Enter

---
## Bước 4 — Duyệt kết nối (Pairing)

Lần đầu chạy Gateway, thiết bị cần được duyệt trước khi hoạt động.

**Xem danh sách đang chờ:**

````bash
openclaw devices list
````

**Duyệt request** — thay `<request-id>` bằng ID từ lệnh trên:

````bash
openclaw devices approve <request-id>
````
